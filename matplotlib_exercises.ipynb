```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (8, 6))
x = range(-100, 100)
equation = [x**2 - x + 2 for x in x]

plt.plot(x, equation)
plt.annotate('(0,0)', xy= (0, 100), xytext = (0, 2000), arrowprops = {'facecolor': 'orange'})
```

```python
rangelist = []
for x in range(1, 1000):
    rangelist.append(x*.01)

y1 = [x ** (1/2) for x in rangelist]
y2 = [x ** 3 for x in rangelist]
y3 = [2 ** x for x in rangelist]
y4 = [(1/(x+1)) for x in rangelist]

plt.plot(rangelist, y1)
plt.title("$\sqrt{x}$", fontsize = 14);
```

```python
plt.plot(rangelist, y2)
plt.title("$x^3$", fontsize = 14);
```

```python
plt.plot(rangelist, y3)
plt.title("$2^x$", fontsize = 14);
```

```python
plt.plot(rangelist, y4)

plt.title("$1/(x+1)$", fontsize = 14);
```

```python
plt.figure(figsize = (10, 8))

plt.subplot(2, 2, 1)
plt.plot(rangelist, y1)
plt.title("$\sqrt{x}$", fontsize = 14);

plt.subplot(2, 2, 2)
plt.plot(rangelist, y2)
plt.title("$x^3$", fontsize = 14);

plt.subplot(2, 2, 3)
plt.plot(rangelist, y3)
plt.title("$2^x$", fontsize = 14);

plt.subplot(2, 2, 4)
plt.plot(rangelist, y4)

plt.title("$1/(x+1)$", fontsize = 14);
```

```python
plt.figure(figsize = (10, 8))

plt.plot(rangelist, y1, label = "$\sqrt{x}$")
plt.plot(rangelist, y2, label = "$x^3$")
plt.plot(rangelist, y3, label = "$2^x$")
plt.plot(rangelist, y4, label = "$1/(x+1)$")
plt.ylim(0, 50)
plt.xlim(0, 10)
plt.legend(fontsize = 'x-large')
plt.title("Various math equations ", fontsize = "x-large");
```
