```python
import matplotlib.pyplot as plt
import math as m
```

```python
plt.figure(figsize = (16, 9))
range_list = []
for x in range(0,100):
    range_list.append(x *.1)
log_range_list = []

for x in range(1, 1000):
    log_range_list.append(x*.01)



y1 = [1 for x in range_list]
y2 = [m.log(x) for x in log_range_list]
y3 = [x for x in range_list]
y4 = [x * m.log(x) for x in log_range_list]
y5 = [x ** 2 for x in range_list]
y6 = [2 ** x for x in range_list]
y7 = [m.gamma(x + 1) for x in range_list]
y8 = [x ** x for x in range_list]


plt.plot(range_list, y1, label = "$O(1)$")
plt.plot(log_range_list, y2, label = "$O(log n)$")
plt.plot(range_list, y3, label = "$O(n)$")
plt.plot(log_range_list, y4, label = "$O(n log n)$")
plt.plot(range_list, y5, label = "$O(n^2)$")
plt.plot(range_list, y6, label = "$O(2^n)$", c = 'lime')
plt.plot(range_list, y7, label = "$O(n!)$")
plt.plot(range_list, y8, label = "$O(n^n)$")

plt.title("Big O Notation")
plt.xlabel("Elements")
plt.ylabel("Operations")
plt.legend(fontsize = 'large')
plt.ylim(-5,15)
plt.xlim(0,10)
```
