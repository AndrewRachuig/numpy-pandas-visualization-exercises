```python
from pydataset import data
import pandas as pd
import numpy as np
```

```python
# 1. Copy the code from the lesson to create a dataframe full of student grades.


np.random.seed(123)

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

# randomly generate scores for each student for each subject
# note that all the values need to have the same length here
math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))

df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})

df
```

```python
# 1. a. Create a column named passing_english that indicates whether each student has a passing grade in english.
df['passing_english'] = df.english > 70
df
```

```python
#1. b. Sort the english grades by the passing_english column. How are duplicates handled?
df.sort_values(by="passing_english")
# Duplicates are simply displayed one after another, by ascending index
```

```python
#1. c. Sort the english grades first by passing_english and then by student name. All the students that are failing english should be first, and within the students that are failing english they should be ordered alphabetically. The same should be true for the students passing english. (Hint: you can pass a list to the .sort_values method)
df.sort_values(by= ["passing_english", "name"])
```

```python
#1. d. Sort the english grades first by passing_english, and then by the actual english grade, similar to how we did in the last step.
df.sort_values(by=['passing_english', 'english'])
```

```python
#1. e. Calculate each students overall grade and add it as a column on the dataframe. The overall grade is the average of the math, english, and reading grades.
df['overall_grade_avg'] = (df.math + df.english + df.reading) / 3
df
```

```python
# 2. Load the mpg dataset. Read the documentation for the dataset and use it for the following questions:
mpg = data('mpg')
data('mpg', show_doc=True)
```

```python
# How many rows and columns are there?
mpg.shape

#234 rows and 11 columns
```

```python
# What are the data types of each column?
mpg.dtypes

# manufacturer     object
# model            object
# displ           float64
# year              int64
# cyl               int64
# trans            object
# drv              object
# cty               int64
# hwy               int64
# fl               object
# class            object
```

```python
# Summarize the dataframe with .info and .describe
mpg.info(), mpg.describe()
```

```python
# Rename the cty column to city. Rename the hwy column to highway.

mpg = mpg.rename(columns = {'cty':'city', 'hwy':'highway'})
mpg
```

```python
# Do any cars have better city mileage than highway mileage?
(mpg.city - mpg.highway).sort_values(ascending = False)

#No
```

```python
# Create a column named mileage_difference this column should contain the difference between highway and city mileage for each car.
mpg['mileage_difference'] = (mpg.highway - mpg.city)
mpg
```

```python
# Which car (or cars) has the highest mileage difference?
mpg.sort_values(by="mileage_difference", ascending = False).head()

# The 2008 honda civic and the 1999 volkswagen new beetle both have the highest difference at 12.
```

```python
# Which compact class car has the lowest highway mileage? The best?
mpg[(mpg['class'] == 'compact')].sort_values(by='highway').head()
#The 1999 volkswagen jetta with a 2.8 liter 6 cylinder automatic transmission has the worst highway mileage for a compact car

mpg[(mpg['class'] == 'compact')].sort_values(by='highway', ascending = False).head()
# The 1999 volkwagen jetta with a 1.9 liter 4 cylinder manual transmission has the best highway mileage for a compact car
```

```python
# Create a column named average_mileage that is the mean of the city and highway mileage.
mpg['average_mileage'] = (mpg.city + mpg.highway)/2
mpg
```

```python
# Which dodge car has the best average mileage? The worst?
mpg[mpg['manufacturer'] == 'dodge'].sort_values(by='average_mileage', ascending = False).head()
# The 1999 dodge caravan 2wd has the best average mileage for a dodge vehicle.

mpg[mpg['manufacturer'] == 'dodge'].sort_values(by='average_mileage').head()
# There is a four way tie for worst average mileage for a dodge vehicle. The following all averaged 10.5 mpg:
# The 2008 ram 1500 pickup 4wd with manual transmission, the 2008 ram 1500 pickup 4wd with automatic transmission,
# the durango 4wd, and the dakota pickup 4wd
```

```python
# 3. Load the Mammals dataset. Read the documentation for it, and use the data to answer these questions:
mammals = data('Mammals')
data('Mammals', show_doc=True)
```

```python
# How many rows and columns are there?
mammals.shape

# 107 rows and 4 columns
```

```python
# What are the data types?
mammals.dtypes
```

```python
# Summarize the dataframe with .info and .describe
mammals.info(), mammals.describe()
```

```python
# What is the the weight of the fastest animal?
mammals.loc[mammals.speed.idxmax()].weight

#alternative without using .loc
# mammals[mammals.speed.isin(mammals.speed.nlargest(1))].weight.values[0]
```

```python
# What is the overal percentage of specials?
mammals[mammals.specials].shape[0]

(mammals[mammals.specials].shape[0]) / mammals.shape[0] * 100

# 9.35% of all mammals are specials in the given list.
```

```python
# How many animals are hoppers that are above the median speed? 
mammals[(mammals.hoppers) & (mammals.speed > mammals.speed.median())]

# What percentage is this?
(mammals[(mammals.hoppers) & (mammals.speed > mammals.speed.median())].shape[0])/ mammals.shape[0] * 100

# This is 6.54% of all mammals in the given list.
```
