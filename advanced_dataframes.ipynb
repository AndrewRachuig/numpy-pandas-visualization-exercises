```python
# 1. Copy the users and roles DataFrames from the examples above.
users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['bob', 'joe', 'sally', 'adam', 'jane', 'mike'],
    'role_id': [1, 2, 3, 3, np.nan, np.nan]
})


roles = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})
roles 
```

```python
# 2. What is the result of using a right join on the DataFrames?

users.merge(roles, left_on='role_id', right_on='id', how="right", indicator=True)
```

```python
# 3. What is the result of using an outer join on the DataFrames?

users.merge(roles, left_on='role_id', right_on='id', how="outer", indicator=True)
```

```python
# 4. What happens if you drop the foreign keys from the DataFrames and try to merge them?
fk_dropped = users.drop(columns='role_id')

# It depends on what kind of join you do, if you do an outer join, it gives you back a array of what data it can but it's pretty useless
fk_dropped.merge(roles, how='outer', indicator=True)

# If you do an inner join it gives you back nothing as it doesn't know how to match the data
fk_dropped.merge(roles, how='inner', indicator=True)
```

```python
# 13. Create a column named average_mileage like you did in the DataFrames exercises; this is the mean of the city and highway mileage.
mpg['average_mileage'] = (mpg.cty + mpg.hwy)/2
mpg
```

```python
# 14. Create a new column on the mpg dataset named is_automatic that holds boolean values denoting whether the car has an automatic transmission.
mpg['is_automatic'] = mpg.trans.apply(lambda x: True if 'auto' in x else False)

mpg
```

```python
# 15. Using the mpg dataset, find out which which manufacturer has the best miles per gallon on average?
manuf_ave_mpg = mpg.groupby('manufacturer').agg({'average_mileage': 'mean'})
pd.DataFrame(manuf_ave_mpg.average_mileage.nlargest(1))
```

```python
# 16. Do automatic or manual cars have better miles per gallon?

mpg.groupby('is_automatic')[["average_mileage"]].agg('mean')
```

```python
# 7. Write the code necessary to create a cross tabulation of the number of titles by department. 
# (Hint: this will involve a combination of SQL code to pull the necessary data and python/pandas code to perform the manipulations.)
url = env.get_db_url('employees')


query = '''
SELECT departments.dept_name, titles.title
FROM employees
JOIN dept_emp USING (emp_no)
JOIN titles USING (emp_no)
JOIN departments on departments.dept_no = dept_emp.dept_no;
'''

dept_with_titles = pd.read_sql(query, url)
```

```python
pd.crosstab(dept_with_titles.dept_name, dept_with_titles.title)
```
